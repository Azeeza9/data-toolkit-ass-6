6)What are the differences between NumPy arrays and Python lists?
- NumPy arrays and Python lists can both store sequences of data, they are very different in terms of performance, functionality, and memory usage.
  - 1) Data Type Consistency:
NumPy arrays require that all elements are of the same data type (e.g., all integers or all floats). This allows for more efficient storage and computation.
Python lists can contain elements of different data types—you can mix integers, strings, floats, even other lists.
  - 2)Performance:
NumPy arrays are much faster than Python lists for numerical operations. This is because NumPy is implemented in C and supports vectorized operations, meaning it can process entire arrays at once.Python lists are slower because operations typically require loops, and each element is a separate Python object, which adds overhead.
  - 3)Memory Efficiency:
NumPy arrays use less memory because the data is stored in a contiguous block in memory and all elements are the same type.Python lists are more memory-hungry because each element is stored as a full Python object, with its own metadata.
  - 4)Functionality for Math and Science:
NumPy arrays come with built-in mathematical functions and support for advanced operations like matrix multiplication, linear algebra, Fourier transforms, etc.
Python lists don’t have built-in support for such operations; you would need to write loops or use other libraries.

15)How does NumPy handle multidimensional arrays?
- NumPy handles multidimensional arrays through its ndarray (n-dimensional array) object, which is a powerful and efficient container for large data sets of homogeneous types (all elements have the same data type).
  - 1)Structure of ndarray:
Shape: A tuple of integers indicating the size of the array along each dimension. For example, (3, 4, 5) means the array has 3 "blocks," each containing 4 rows and 5 columns.Dimensions (ndim): The number of axes or dimensions. A 2D array has 2 dimensions, a 3D array has 3, and so on.Data type (dtype): The type of the elements, e.g., int32, float64.Data buffer: A contiguous block of memory holding the array elements.
  - 2)Indexing and Slicing:
Supports multidimensional indexing, e.g., arr[2, 3] for a 2D array.Slicing works across multiple dimensions, e.g., arr[:, 1:4] slices rows or columns as needed.
  - 3)Memory Layout:
Arrays can be stored in row-major (C-style) or column-major (Fortran-style) order.Internally, NumPy stores the array as a contiguous block of memory (or strided views), which makes operations fast and efficient.
  - 4)Broadcasting:
NumPy can automatically expand arrays with smaller dimensions to perform element-wise operations with larger arrays, based on broadcasting rules.
  - 5)Shape Manipulation:
Functions like .reshape(), .transpose(), .swapaxes(), etc., allow changing the shape or the order of axes without copying the data.

```python
#4)How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#9)Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [85, 42, 77, 90, 66]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

threshold = 70

filtered_df = df[df['Score'] > threshold]

print("\nFiltered DataFrame (Score > 70):")
print(filtered_df)

```
